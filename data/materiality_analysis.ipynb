# Materiality Map Draft 데이터 분석

본 노트북은 `data/materiality_map_draft.csv`의 데이터 분포와 업종별 중요도 매핑을 시각화 및 집계하기 위한 분석 코드입니다.

```python
import pandas as pd
import os

# 1. 데이터 로드
# 파일 경로가 data 폴더 내에 있는지, 상위 폴더에 있는지 체크하여 로드
csv_path = 'materiality_map_draft.csv' if os.path.exists('materiality_map_draft.csv') else '../data/materiality_map_draft.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
# 2. is_material 빈도 분석
print("--- 중요도(is_material) 전체 빈도수 ---")
print(df['is_material'].value_counts())

print("\n--- 비율 (%)) ---")
print(df['is_material'].value_counts(normalize=True) * 100)
```

```python
# 3. 뉴스 카테고리별 중요(1) 매핑 집계
print("--- 뉴스 카테고리별 중요도 집계 ---")
category_agg = df.groupby('news_category')['is_material'].agg(['count', 'sum'])
category_agg.columns = ['전체 업종 수', '중요(1) 판정 업종 수']
category_agg['비중요(0) 판정 업종 수'] = category_agg['전체 업종 수'] - category_agg['중요(1) 판정 업종 수']
category_agg
```

```python
# 4. 업종(sasb_sector)별 중요도 집계
print("--- 업종별 중요 카테고리 개수 ---")
sector_agg = df.groupby('sasb_sector')['is_material'].agg(['count', 'sum'])
sector_agg.columns = ['전체 카테고리 수', '중요(1) 카테고리 수']
sector_agg
```
